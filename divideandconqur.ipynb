```python
#Find min and max_val_val_val
def findminandmax(arr,i,j):
    #Small Problem
    if i==j:
        min_val = arr[i]
        max_val = arr[i]
    elif i == j-1:
        if arr[i]<arr[j]:
            min_val = arr[i]
            max_val = arr[j]
        else:
            min_val = arr[j]
            max_val = arr[i]
    else:
        mid = i+(j-i)//2
        min_1,max_1 = findminandmax(arr,i,mid)
        min_2,max_2 = findminandmax(arr,mid+1,j)

        if max_1<max_2:
            max_val = max_2
        else:
            max_val = max_1
        if min_1<min_2:
            min_val = min_1
        else:
            min_val = min_2
    return min_val,max_val
arr = [75,45,95,50,60,67,29,32]
i = 0 
j = len(arr)-1
print(findminandmax(arr,i,j))

```

```python
def power(a,n):
    if n == 0:
        return 1
    elif n == 1:
        return a 
    elif n<0:
        return 1/power(a,-n)
    else:
        min = n//2
        b = power(a,min)
        result = b*b
        if n%2==0:
            return result
        else:
            return result*a
a=4
n=-1
print(power(a,n))
```
