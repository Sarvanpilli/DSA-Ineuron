```python
#Ternary Search 

def ternarySearch(arr,i,j,target):
    while i<=j:
        mid1 = i+(j-i)//3
        mid2 = j-(j-i)//3

        if arr[mid1] == target:
            return mid1
        elif arr[mid2] == target:
            return mid2
        elif target<arr[mid1]:
            return ternarySearch(arr,i,mid1-1,target)
        elif target>arr[mid2]:
            return ternarySearch(arr,mid2+1,j,target)
        else:
            return ternarySearch(arr,mid1+1,mid2-1,target)
    return -1 

arr = [20,25,47,56,59,63,65,79,82]
target = int(input())
i = 0 
j = len(arr)-1
print("Position of the Target element" ,target,"is: ",ternarySearch(arr,i,j,target))


```
