```python
#Ternary Search 

def ternarySearch(arr,i,j,target):
    while i<=j:
        mid1 = i+(j-i)//3
        mid2 = j-(j-i)//3

        if arr[mid1] == target:
            return mid1
        elif arr[mid2] == target:
            return mid2
        elif target<arr[mid1]:
            return ternarySearch(arr,i,mid1-1,target)
        elif target>arr[mid2]:
            return ternarySearch(arr,mid2+1,j,target)
        else:
            return ternarySearch(arr,mid1+1,mid2-1,target)
    return -1 

arr = [20,25,47,56,59,63,65,79,82]
target = int(input())
i = 0 
j = len(arr)-1
print("Position of the Target element" ,target,"is:",ternarySearch(arr,i,j,target))


```

```python
def bubblesort(arr):
    n = len(arr)
    for i in range(n):
        for j in range(0,n-i-1):
            if arr[j]>arr[j+1]:
                arr[j],arr[j+1] = arr[j+1],arr[j]
    return arr
arr = [70,20,50,30,90,5,15]
print("The Sorted Array:",bubblesort(arr))
```

```python
def selection(arr):
    n = len(arr)
    for i in range(n):
        min = i 
        for j in range(i+1,n):
            if arr[j]<arr[min]:
                min = j
        arr[i],arr[min] = arr[min],arr[i]
    return arr
arr = [70,20,50,30,90,5,15]
print("Selection Sort:",selection(arr))

```

```python
#insertion Sort
def insertionsort(arr):
    for i in range(1,len(arr)):
        j = i-1 
        key = arr[i]
        while j>=0 and arr[j]>key:
            arr[j+1] = arr[j]
            j-=1
        arr[j+1] = key
    return arr
arr = [70,65,50,30,90,5,15]
print(insertionsort(arr))
```
