```python
#Reverse 
arr = [int(x) for x in input().split()]
i=0
mid = len(arr)//2
while (i<mid):
    arr[i],arr[len(arr)-i-1]=arr[len(arr)-i-1],arr[i]
    i+=1
print(arr)
```

```python
#minimum 
arr = list(map(int,input().split()))
def minimum(array):
    min = float('inf')
    for i in range(len(arr)):
        if i < min:
            min = array[i]
    return min
print(minimum(arr))
    
```

```python
#maxium
arr = list(map(int,input().split()))
def maxium(array):
    max = float('-inf')
    for i in range(len(arr)):
        if i > max:
            max = arr[i]
    return max 
print(maxium(arr))
```

```python
#binary Search 
arr = list(map(int,input().split()))
l = arr[0]
r = len(arr)
mid = (l+r)//2
target = int(input())
while(l<r):
    if mid[arr] == target :
        print(mid)
    elif mid[arr]>target:
        l = mid+1
    elif mid[arr]<target:
        r = mid - 1

```

```python
#bubble sort 
arr = list(map(int,input().split()))
n = len(arr)
for i in range(n):
    Swapped = False 
    for j in range(0,n-i-1):
        arr[j],arr[j+1]=arr[j+1],arr[j]
        Swapped = True 
    if not Swapped:
        break
print(arr)
```
