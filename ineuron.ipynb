```python
#Arrays Searching 

arr = list(map(int,input().split()))
target = int(input())
found = False
for i in range(len(arr)):
    if arr[i] == target:
        print(i)
        found = True
        break
if not found:
    print("Not found ")
```

```python
#Binary Search 
def binarysearch(arr,left,right):
    
    while left<=right:
        mid = (left+right)//2

        if arr[mid] == target:
            return mid
        elif arr[mid]>target:
            right = mid-1
        else :
            left = mid +1
    return -1 
arr = list(map(int,input().split()))
target = int(input())
left = 0
right = len(arr)-1
print(binarysearch(arr,left,right))
```

```python
#Binary Search using Recursion
def binarysearch(arr,left,right):
    
    while left<=right:
        mid = (left+right)//2

        if arr[mid] == target:
            return mid
        elif arr[mid]>target:
            return binarysearch(arr,left,mid-1)
        else :
            return binarysearch(arr,mid+1,right)
    return -1 
arr = list(map(int,input().split()))
target = int(input())
left = 0
right = len(arr)-1
print(binarysearch(arr,left,right))
```

```python
#Two Sum 
arr = list(map(int,input().split()))
target = int(input())
for i in range(0,len(arr)):
    for j in range(i,len(arr)):
        if arr[i] + arr[j] == target:
            print(i+1,j+1)
```

```python
def point(arr,target):
    left = 0 
    right = len(arr)-1

    while left<=right:
        if left + right == target:
            return left,right
        elif left+right > target:
            right = right - 1 
        else:
            left = left + 1 
    return -1
arr = list(map(int,input().split()))
target = int(input())
print(point(arr,target))
```
